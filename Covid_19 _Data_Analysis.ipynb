Covid-19 Data Analysis


Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the Dataset

```python
df = pd.read_csv("Covid_19_Dataset (2).csv")
df.head()
```

Analyze Daily Trends

```python
daily_cases = df.groupby('Date')['Confirmed'].sum()
```

```python
daily_deaths = df.groupby('Date')['Deaths'].sum()
```

```python
daily_recovered= df.groupby('Date')['Recovered'].sum()
```

Visualization of the Daily Trends

```python
plt.figure(figsize=(10,5))
plt.scatter(daily_cases,daily_deaths)
plt.title("Daily Covid-19 Confirmed cases")
plt.xlabel("Dates")
plt.ylabel("Confirmed cases")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(daily_cases)
plt.title("Daily Covid-19 Confirmed cases")
plt.xlabel("Dates")
plt.ylabel("Confirmed cases")
plt.show()
```

```python
plt.plot(daily_deaths)
plt.title("Daily Covid-19 Death cases")
plt.xlabel("Dates")
plt.ylabel("Death cases")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(daily_recovered)
plt.title("Daily Covid-19 Recovery cases")
plt.xlabel("Dates")
plt.ylabel("Recovery cases")
plt.show()
```
